# 機械学習ライブラリ scikit-learn で単回帰分析

## データを眺める

```python
import pandas as pd
```

```python
df = pd.read_csv('data.csv')
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
plt.scatter(df['x'], df['y'])
plt.show()
```

```python
X = df.iloc[:,0:-1]
y = df.iloc[:,-1]
```

```python
plt.scatter(X, y)
plt.show()
```

## 分析する

```python
from sklearn import linear_model
```

```python
regr = linear_model.LinearRegression()
regr.fit(X, y)
```

```python
a = regr.coef_[0]
b = regr.intercept_
```

```python
plt.scatter(X, y, label='y')
plt.plot(X, a * X + b, label='y_hat')
plt.legend()
plt.show()
```

```python
y_pred = regr.predict(X)
```

```python
plt.scatter(X, y, label='y', color = 'blue')
plt.scatter(X, y_pred, label='y_hat', color = 'red')
plt.legend()
plt.show()
```

## 評価する

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
#　平均二乗誤差
mean_squared_error(y, y_pred)
```

```python
# 決定係数
r2_score(y, y_pred)
```
