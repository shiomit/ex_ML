# 機械学習ライブラリ scikit-learn で重回帰分析

## データを眺める

```python
import pandas as pd
```

```python
df = pd.read_csv('housing.data', header=None, delim_whitespace=True,
                names=['CRIM','ZN','INDUS','CHAS','NOX','RM','AGE','DIS','RAD',
                       'TAX','PTRATIO','B','LSTAT','MEDV'])
df.head()
```

```python
# 散布図行列を書いてみる
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.pairplot(df)
plt.show()
```

```python
sns.pairplot(df.iloc[:,2:8])
plt.show()
```

```python
sns.pairplot(df.loc[:,['RM','DIS','LSTAT','MEDV']])
plt.show()
```

```python
X = df.iloc[:,0:-1]
y = df.iloc[:,-1]
```

```python
X.head()
```

```python
y.head()
```

## 分析する

```python
from sklearn import linear_model
```

```python
regr = linear_model.LinearRegression()
regr.fit(X, y)
```

```python
regr.coef_
```

```python
regr.intercept_
```

```python
y_pred = regr.predict(X)
```

## 評価する

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
#　平均二乗誤差
mean_squared_error(y, y_pred)
```

```python
# 決定係数
r2_score(y, y_pred)
```
