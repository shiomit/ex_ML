# 過学習とは
機械学習において、過度の教師あり学習を行った場合、未学習の問いに対して正しく答えを出力できなくなる現象。
学習のしすぎによって判断の基準が厳しくなるため、少しでもパターンが異なると誤った答えを出力してしまうこと。
オーバーフィッティング。

```python
import pandas as pd
```

```python
df = pd.read_csv('2D_data.csv', header=None, names=['x1','x2','y'])
```

```python
X = df.iloc[:,0:-1]
y = df.iloc[:,-1]
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plotBoundary(X, clf, mesh=True, boundary=True, n_neighbors=1):

    # plot range
    x_min = min(X.iloc[:,0])
    x_max = max(X.iloc[:,0])
    y_min = min(X.iloc[:,1])
    y_max = max(X.iloc[:,1])
    # visualizing decision function
    XX, YY = np.mgrid[x_min:x_max:200j, y_min:y_max:200j] # make a grid
    
    Z = clf.predict(np.c_[XX.ravel(), YY.ravel()]) # evaluate the value 
    
    Z = Z.reshape(XX.shape) # just reshape

    if mesh:
        plt.pcolormesh(XX, YY, Z, shading='auto', zorder=-10) # paint in 2 colors, if Z > 0 or not

    if boundary:
        plt.contour(XX, YY, Z, 
                    colors='k', linestyles='-', levels=[0.5])

    plt.xlim(x_min, x_max)
    plt.ylim(y_min, y_max)
```

```python
from sklearn import svm
```

```python
clf = svm.SVC(kernel='rbf', C=1, gamma=0.5)
```

```python
clf.fit(X, y)
plt.set_cmap(plt.cm.Paired)
plt.scatter(X['x1'], X['x2'], marker='o', s=50, c=y, edgecolor='black')
plotBoundary(X, clf)
plt.show()
```

パラメーターを変更すると

```python
clf = svm.SVC(kernel='rbf', C=1, gamma=3)
```

```python
clf.fit(X, y)
plt.set_cmap(plt.cm.Paired)
plt.scatter(X['x1'], X['x2'], marker='o', s=50, c=y, edgecolor='black')
plotBoundary(X, clf)
plt.show()
```

過学習したモデルは、学習に使ったデータに対してはバッチリ正解できるけど、知らないデータに対しては全然当たらない。
