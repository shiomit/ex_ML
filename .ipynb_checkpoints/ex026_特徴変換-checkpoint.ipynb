# 特徴変換（PCA）
特徴選択は、特徴量の中から効果的なものだけを選ぶ方法でした。  
今度は、特徴量に何らかの変換を行う手法を２つ説明します。  
１．主成分分析（Principal Component Analysis）を使って特徴量を次元削減する方法  
２．多項式変換  
これらのうち、まずはPCAから説明します。

```python
import pandas as pd
df = pd.read_csv('wdbc.data', header=None,
                 names=['id','diagnosis',
                        'radius_mean','texture_mean','perimeter_mean','area_mean','smoothness_mean',
                        'compactness_mean','concavity_mean','concave points_mean','symmetry_mean','fractal_dimension_mean',
                        'radius_se','texture_se','perimeter_se','area_se','smoothness_se',
                        'compactness_se','concavity_se','concave points_se','symmetry_se','fractal_dimension_se',
                        'radius_worst','texture_worst','perimeter_worst','area_worst','smoothness_worst',
                        'compactness_worst','concavity_worst','concave points_worst','symmetry_worst','fractal_dimension_worst'])
X = df.iloc[:,2:] # 機械学習では、このXを「特徴量」と呼びます。
y = df.iloc[:,1].replace({'M':0, 'B':1})
```

```python
# 散布図行列を書いてみましょう。
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 全30項目のプロット sns.pairplot(X) は重すぎるので10項目だけ表示してみます。
sns.pairplot(X.iloc[:,0:10])
plt.show()
```

```python
# この中からradius_meanとperimeter_meanだけをプロットしてみましょう。

sns.pairplot(X.iloc[:,[0,2]])
plt.show()

# ほとんど一直線の関係になっています。
# つまりradius（半径）がわかればperimeter（周囲長）がわかる。反対に、perimeterがわかればradiusがわかります。
# ということは、radiusとperimeterの両方は必要なく、どちらか片方だけあればデータとしては十分だということです。
```

## PCAを体験

```python
# 特徴量Xから0番目と2番目の特徴を取り出して、同じスケールでプロットします。
X2 = X.iloc[:, [0, 2]]
plt.scatter(X2.iloc[:, 0], X2.iloc[:, 1])
plt.xlim(0, 200)
plt.ylim(0, 200)
plt.show()
```

```python
# PCAで特徴量を変換します。
from sklearn.decomposition import PCA
pca = PCA()
pca.fit(X2)
X2_pca = pca.transform(X2)
plt.scatter(X2_pca[:, 0], X2_pca[:, 1])
plt.xlim(-60, 120)
plt.ylim(-60, 120)
plt.show()
# 2次元のデータが1次元のデータに削減されました。
```

```python
# もう一つの例
# 今度は特徴量Xから6番目と7番目の特徴を取り出して、同じスケールでプロットします。
X3 = X.iloc[:, [6, 7]]
plt.scatter(X3.iloc[:, 0], X3.iloc[:, 1])
plt.xlim(0, 0.5)
plt.ylim(0, 0.5)
plt.show()
```

```python
# PCAで特徴量を変換します。
pca = PCA()
pca.fit(X3)
X3_pca = pca.transform(X3)
plt.scatter(X3_pca[:, 0], X3_pca[:, 1])
plt.xlim(-0.1, 0.4)
plt.ylim(-0.25, 0.25)
plt.show()
# 一つ目の例ほど顕著ではありませんが、横方向のばらつきに比べて縦方向のばらつきは極端に少なく
# データの変動の大方が横方向だけで説明できると考えてください。
```

## 特徴量をPCAで変換して機械学習を行う

```python
# データを訓練用とテスト用に分ける
from sklearn.model_selection import ShuffleSplit
ss = ShuffleSplit(n_splits=1, train_size=0.8, test_size=0.2, random_state=0)
train_index, test_index = next(ss.split(X, y))
X_train, X_test = X.iloc[train_index], X.iloc[test_index]
y_train, y_test = y.iloc[train_index], y.iloc[test_index]
```

```python
# 特徴量を標準化する
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
X_train_scale = scaler.transform(X_train)
X_test_scale = scaler.transform(X_test)
```

```python
# 標準化した特徴量をPCAで変換
pca = PCA()
pca.fit(X_train_scale)
X_train_scale_pca = pca.transform(X_train_scale)
X_test_scale_pca  = pca.transform(X_test_scale)
```

```python
# 変換した特徴量で機械学習する（パラメーターCを変えながら）
from sklearn import linear_model
import numpy as np
from sklearn.model_selection import cross_val_score

clf = linear_model.LogisticRegression()
C_range_exp = np.linspace(start=-30, stop=10, num=41)
C_range = 10 ** C_range_exp
all_scores_mean = []
all_scores_std = []
for C in C_range:
    clf.C = C
    scores = cross_val_score(clf, X_train_scale_pca[:, 0:3], y_train, cv=10) # 変換した特徴量の先頭3次元だけで学習する
    all_scores_mean.append(scores.mean())
    all_scores_std.append(scores.std())
```

```python
# パラメーターごとのスコアを図示する
import matplotlib.pyplot as plt
plt.plot(C_range_exp, all_scores_mean)
plt.errorbar(C_range_exp, all_scores_mean, yerr=all_scores_std)
plt.ylim(0, 1.2)
plt.ylabel('accuracy')
plt.xlabel('$\log_{10}$(C)')
plt.title('Accuracy for different values of C')
plt.show()
```

```python
# 後でnp.argmaxを使いたいので、リストからnumpyアレイに変換しておく。
all_scores_mean = np.array(all_scores_mean)
all_scores_std = np.array(all_scores_std)
```

```python
all_scores_mean.max()
```

```python
max_index = np.argmax(all_scores_mean)
max_index
```

```python
# 最適なパラメーターをテストデータに適用する
clf.C = 10 ** C_range_exp[max_index]
clf.fit(X_train_scale, y_train)
clf.score(X_test_scale, y_test)
```

## PCAまとめ
このデータの場合、元の特徴量は30次元ありましたが、PCAで3次元に削減しても、そこそこよいスコアが得られました。
