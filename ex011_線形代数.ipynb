# ベクトルで微分
$$
\boldsymbol{c}=\left[
 \begin{array}{r}
 3 \\
 4
 \end{array}
\right],
\quad
\boldsymbol{x}=\left[
 \begin{array}{r}
 x_1 \\
 x_2
 \end{array}
\right]
$$

$$
\boldsymbol{c}^{\mathrm{T}}\boldsymbol{x}=\left[
 \begin{array}{rr}
 3 & 4
 \end{array}
\right]
\left[
 \begin{array}{r}
 x_1 \\
 x_2
 \end{array}
\right]=3x_1+4x_2
$$

$$
\frac{\partial}{\partial{\boldsymbol{x}}}\boldsymbol{c}^{\mathrm{T}}\boldsymbol{x}=\left[
 \begin{array}{r}
 \frac{\partial}{\partial{\boldsymbol{x}_1}}(3x_1+4x_2) \\
 \frac{\partial}{\partial{\boldsymbol{x}_2}}(3x_1+4x_2)
 \end{array}
\right]
$$

\begin{eqnarray}
\frac{\partial}{\partial{\boldsymbol{x}_1}}(3x_1+4x_2)&=&\frac{\partial}{\partial{\boldsymbol{x}_1}}(3x_1)+\frac{\partial}{\partial{\boldsymbol{x}_1}}(4x_2) \\
& = & 3\frac{\partial}{\partial{\boldsymbol{x}_1}}(x_1)+4\frac{\partial}{\partial{\boldsymbol{x}_1}}(x_2) \\
& = & 3
\end{eqnarray}

\begin{eqnarray}
\frac{\partial}{\partial{\boldsymbol{x}_2}}(3x_1+4x_2)&=&\frac{\partial}{\partial{\boldsymbol{x}_2}}(3x_1)+\frac{\partial}{\partial{\boldsymbol{x}_2}}(4x_2) \\
& = & 3\frac{\partial}{\partial{\boldsymbol{x}_2}}(x_1)+4\frac{\partial}{\partial{\boldsymbol{x}_2}}(x_2) \\
& = & 4
\end{eqnarray}

$$
\therefore
\quad
\frac{\partial}{\partial{\boldsymbol{x}}}\boldsymbol{c}^{\mathrm{T}}\boldsymbol{x}=\left[
 \begin{array}{r}
 3 \\
 4
 \end{array}
\right]
$$
