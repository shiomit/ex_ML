# 特徴量の標準化

## データの準備
ex018で使ったUCI Machine Learning Repositoryの wdbc.dataを再利用します。

```python
import pandas as pd
```

```python
df = pd.read_csv('wdbc.data', header=None,
                 names=['id','diagnosis',
                        'radius_mean','texture_mean','perimeter_mean','area_mean',
                        'smoothness_mean','compactness_mean','concavity_mean',
                        'concave points_meen','symmetry_mean','fractal_dimension_mean',
                        'radius_se','texture_se','perimeter_se','area_se','smoothness_se',
                        'compactness_se','concavity_se','concave points_se','symmetry_se',
                        'fractal_dimension_se','radius_worst','texture_worst',
                        'perimeter_worst','area_worst','smoothness_worst',
                        'compactness_worst','concavity_worst','concave points_worst',
                        'symmetry_worst','fractal_dimension_worst'])
```

```python
df.head()
```

```python
X = df.iloc[:,2:] # 機械学習では、このXを「特徴量」と呼びます。
```

```python
y = df.iloc[:,1].replace({'M':0, 'B':1})
```

## 特徴量の分布をみる

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(X.iloc[:,3], X.iloc[:,4])
plt.xlabel('area_mean')
plt.ylabel('smoothness_mean')
plt.show()
```

## 縦横のスケールを同じにすると

```python
plt.scatter(X.iloc[:,3], X.iloc[:,4])
plt.xlim(0, 3000)
plt.ylim(0, 3000)
plt.xlabel('area_mean')
plt.ylabel('smoothness_mean')
plt.show()
```

## 特徴量を標準化しないまま機械学習
データを訓練用とテスト用に分け、訓練データでクロスバリデーションして、テストデータで評価する。

```python
from sklearn.model_selection import ShuffleSplit
ss = ShuffleSplit(n_splits=1, train_size=0.8, test_size=0.2, random_state=0)
train_index, test_index = next(ss.split(X, y))
X_train, X_test = X.iloc[train_index], X.iloc[test_index]
y_train, y_test = y.iloc[train_index], y.iloc[test_index]
```

```python
from sklearn import linear_model
clf = linear_model.LogisticRegression()
```

```python
import numpy as np
from sklearn.model_selection import cross_val_score
C_range_exp = np.linspace(start=-30, stop=10, num=41)
C_range = 10 ** C_range_exp
all_scores_mean = []
all_scores_std = []
for C in C_range:
    clf.C = C
    scores = cross_val_score(clf, X_train, y_train, cv=10)
    all_scores_mean.append(scores.mean())
    all_scores_std.append(scores.std())
```

```python
import matplotlib.pyplot as plt
plt.plot(C_range_exp, all_scores_mean)
plt.errorbar(C_range_exp, all_scores_mean, yerr=all_scores_std)
plt.ylim(0, 1.2)
plt.ylabel('accuracy')
plt.xlabel('$\log_{10}$(C)')
plt.title('Accuracy for different values of C')
plt.show()
```

```python
# 後でnp.argmaxを使いたいので、リストからnumpyアレイに変換しておく。
all_scores_mean = np.array(all_scores_mean)
all_scores_std = np.array(all_scores_std)
```

```python
all_scores_mean.max()
```

```python
max_index = np.argmax(all_scores_mean)
```

```python
max_index
```

```python
# 最適なパラメーターをテストデータに適用する
clf.C = 10 ** C_range_exp[max_index]
clf.fit(X_train, y_train)
clf.score(X_test, y_test)
```

## データの標準化
データ$X$に対し
$$
\frac{X-\mu}{\sigma}
$$
と変換すること。  
ただし  
$\mu$は$X$の平均  
$\sigma$は$X$の標準偏差  

## 特徴量を標準化する

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train_scale = scaler.transform(X_train)
X_test_scale = scaler.transform(X_test)
```

```python
plt.scatter(X_train_scale[:,3], X_train_scale[:,4], c='blue', label='train')
plt.scatter(X_test_scale[:,3], X_test_scale[:,4], c='red', label='test')
plt.xlabel('area_mean')
plt.ylabel('smoothness_mean')
plt.legend()
plt.show()
```

## 標準化した特徴量で機械学習

```python
C_range_exp = np.linspace(start=-30, stop=10, num=41)
C_range = 10 ** C_range_exp
all_scores_mean = []
all_scores_std = []
for C in C_range:
    clf.C = C
    scores = cross_val_score(clf, X_train_scale, y_train, cv=10)
    all_scores_mean.append(scores.mean())
    all_scores_std.append(scores.std())
```

```python
import matplotlib.pyplot as plt
plt.plot(C_range_exp, all_scores_mean)
plt.errorbar(C_range_exp, all_scores_mean, yerr=all_scores_std)
plt.ylim(0, 1.2)
plt.ylabel('accuracy')
plt.xlabel('$\log_{10}$(C)')
plt.title('Accuracy for different values of C')
plt.show()
```

```python
# 後でnp.argmaxを使いたいので、リストからnumpyアレイに変換しておく。
all_scores_mean = np.array(all_scores_mean)
all_scores_std = np.array(all_scores_std)
```

```python
all_scores_mean.max()
```

```python
max_index = np.argmax(all_scores_mean)
```

```python
max_index
```

```python
# 最適なパラメーターをテストデータに適用する
clf.C = 10 ** C_range_exp[max_index]
clf.fit(X_train_scale, y_train)
clf.score(X_test_scale, y_test)
```

## 結果について
標準化していないデータでのテスト結果 94.7%  
標準化したデータでのテスト結果 96.5%
